```python
import pandas as pd

df = pd.read_csv("MiniProject.csv")
df.head()
```

```python
helpfulness_to_visits = pd.crosstab(df.helpfulness, df.visits)
helpfulness_to_visits
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ct = pd.crosstab(df['helpfulness'], df['visits'])
ct = ct.reindex([
    "Helpful",
    "Very Helpful",
    "Neither Helpful nor Unhelpful",
    "Unhelpful",
    "N/A, never interacted with SRC staff"
], axis=0)
ct = ct[[
    "0, I don't go to the SRC",
    "1-2 times",
    "3-4 times",
    "5 or more times"
]]

plt.figure(figsize=(8, 8))
sns.heatmap(ct)
plt.title("Joint Distribution of Helpfulness and Visits")
plt.xlabel("Visits")
plt.ylabel("Helpfulness")
plt.show()
```
