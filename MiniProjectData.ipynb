```python
#crosstab is a pivottable that holds counts
visit_to_satisfaction = pd.crosstab(df.visits, df.satisfaction)
visit_to_satisfaction 
```

```python
satisfaction_to_visits = pd.crosstab(df.satisfaction, df.visits)
satisfaction_to_visits
```

```python
energy_to_visits = pd.crosstab(df.energy, df.visits)
energy_to_visits
```

```python
standing_to_visits = pd.crosstab(df.standing, df.visits)
standing_to_visits
```

```python
#this is for counts
visits_to_energy = pd.crosstab(df.visits, df.energy)
visits_to_energy
```

```python
#this for the mean of visits to energy levels

visits_to_energy_mean = pd.pivot_table(df, index = "visits", values = "energy", aggfunc = "mean")
visits_to_energy_mean 
```

```python
#standing to energy level count 
standing_to_energy = pd.crosstab(df.standing, df.energy)
standing_to_energy
```

```python
standing_to_energy_mean = pd.pivot_table(df, index = "standing", values = "energy", aggfunc = "mean")
standing_to_energy_mean
```

```python
#this is for standing to diet level
standing_to_diet = pd.crosstab(df.standing, df.diet)
standing_to_diet 

```
